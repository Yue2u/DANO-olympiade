```python
import pandas as pd
```

```python
kick = pd.read_csv("kickstarter.csv")
```

```python
len(kick[(kick['cont'] == 0) & (kick["n_img"] == 0)])
```

```python
len_cont = len(kick[kick["cont"] == 0])
len_cont
```

```python
len_img = len(kick[kick["n_img"] == 0])
len_img
```

```python
num_cont = []
for i in range(0, len_cont):
  num_cont.append(kick[kick['cont'] == 0].iloc[i][0])
```

```python
num_img = []
for k in range(0, len_img):
  num_img.append(kick[kick['n_img'] == 0].iloc[k][0])
```

```python
common = 0
num_iskl = []
iskl = []
for m in range(0, len_cont):
  if num_cont[m] in num_img:
    common += 1
  else:
    num_iskl.append(num_cont[m])
    iskl.append(m)
```

```python
kick['success_rate'] = kick['pled']/kick['goal']
```

```python
from matplotlib import pyplot as plt
import numpy as np
kick = kick[kick['goal'] > 5]
kick = kick[kick['period'] < 80]
kick = kick[kick['n_vid'] < 10]

yx = []

for i in range(12):
    yx.append(kick[kick['created'] == i]['success_rate'].mean())

plt.figure(figsize=(5, 5)) # устанавливаем размер графика - 5 * 5 дюймов

plt.xlabel("количество проектов")
# подписываем ось Х
plt.ylabel("средний success_rate")
# подписываем ось У

plt.plot(np.arange(12), yx)
plt.show()
```
