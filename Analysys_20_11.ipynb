```python
cols = ['rpg', 'platformer', 'shooter', 'fighting', 'survival', 'horror',
       'strategy', 'arcade', 'simulator', 'mmo', 'indie', 'action', 'quest',
       'adventure']


l = set()
for col1 in cols:
    for col2 in cols:
        for col3 in cols:
            if col1 != col2 and col2 != col3 and col1 != col3 and data[(data[col1] == 1) & (data[col2] == 1) & (data[col3] == 1)].shape[0] > 9:
                percentage = data[(data[col1] == 1) & (data[col2] == 1) & (data[col3] == 1) & (data['success_rate'] >= 1)].shape[0] / data[(data[col1] == 1) & (data[col2] == 1) & (data[col3] == 1)].shape[0]
                col1, col2, col3 = sorted([col1, col2, col3])
                l.add((col1, col2, col3, percentage, data[(data[col1] == 1) & (data[col2] == 1) & (data[col3] == 1)].shape[0]))

l = list(l)
l.sort(key=lambda x: x[3])
l
```
