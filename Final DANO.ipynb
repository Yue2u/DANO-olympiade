```python
import pandas as pd
```

```python
kick = pd.read_csv("kickstarter.csv")
```

```python
len(kick[(kick['cont'] == 0) & (kick["n_img"] == 0)])
```

```python
len_cont = len(kick[kick["cont"] == 0])
len_cont
```

```python
len_img = len(kick[kick["n_img"] == 0])
len_img
```

```python
num_cont = []
for i in range(0, len_cont):
  num_cont.append(kick[kick['cont'] == 0].iloc[i][0])
```

```python
num_img = []
for k in range(0, len_img):
  num_img.append(kick[kick['n_img'] == 0].iloc[k][0])
```

```python
common = 0
num_iskl = []
iskl = []
for m in range(0, len_cont):
  if num_cont[m] in num_img:
    common += 1
  else:
    num_iskl.append(num_cont[m])
    iskl.append(m)
```
